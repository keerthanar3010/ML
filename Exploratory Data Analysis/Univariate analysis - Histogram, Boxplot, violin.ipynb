# Univariate analysis

Univariate analysis : One variable/feature analysis wrt to target

Univariate analysis is a statistical technique that involves the examination of a single variable to understand its distribution, central tendency, variability, and other characteristics. In the context of machine learning and deep learning, univariate analysis is crucial for data preprocessing, feature selection, and gaining insights into the dataset.




Importance of Univariate Analysis in Machine Learning and Deep Learning :

Data Understanding:
Univariate analysis helps in understanding the nature of the data, which is crucial for making informed decisions about preprocessing, feature engineering, and model selection.

Feature Selection:
By analyzing individual features, practitioners can identify which features are relevant and which may be redundant or irrelevant, aiding in feature selection and dimensionality reduction.

Data Preprocessing:
Insights from univariate analysis can guide data preprocessing steps, such as normalization, transformation, or handling missing values.

Model Performance:
Understanding the distribution of target variables is essential for selecting appropriate algorithms and evaluation metrics. For instance, regression models may assume normally distributed errors.

Outlier Treatment:
Identifying and addressing outliers can improve model robustness and performance, as outliers can skew results and lead to overfitting.

Assumption Checking:
Many machine learning algorithms have underlying assumptions about the data (e.g., linearity, normality). Univariate analysis helps in checking these assumptions.

```python
# Import Iris dataset

!pip install matplotlib pandas numpy seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris

#load the dataset
iris = load_iris()

# Convert to pandas DataFrame
iris_df = pd.DataFrame(data=iris.data,columns=iris.feature_names)

# Add the target column
iris_df['target'] = iris.target
```

# Method 1 : Descriptive statistic -- "describe()" function

```python
# Descriptive statistic analysis of Numerical variable using "describe()" function

iris_df.describe()
```

# Method 2 : Histogram

```python
# Histogram : Shows the frequency distribution of the data.

sns.histplot(data=iris_df, x="sepal length (cm)", hue='target',kde=True, palette='Set2', bins=15)
plt.show()


# data=iris_df: This specifies that the entire DataFrame is being used for the plot.
# x="sepal length (cm)": This specifies the column for the x-axis.
# hue="target": This specifies the column that will be used to color the histogram bars based on different categories.
# kde=True, provides a smooth curve that represents the PDF of the variable, making it easier to visualize the distribution of the data.
# A kernel density plot (KDE) is a visualization technique that shows the distribution of data in a dataset using a continuous probability density curve. It's also known as a density plot or density trace graph
# palette='Set2',The palette parameter specifies the color palette to be used for the different categories when using the hue parameter.
# The 'Set2' palette is a predefined color palette in Seaborn 
# set bins=15, the range of the data will be divided into 15 equal-width intervals, and the histogram will show the frequency of data points that fall into each of these intervals.
```

```python
# Histogram without KDE (kernel density curve)

sns.histplot(data=iris_df, x="sepal length (cm)", hue='target',palette='Set2', bins=15)
plt.show()
```

# Method 3 : Boxplot

Key Elements of a Box Plot 

a. Median (50%):
The middle line inside the box represents the median (50th percentile or Q2)
It is the middle value of the dataset when arranged in ascending order.
50% of the data points are below this value, and 50% are above.

b. First Quartile (25%):
The lower edge of the box represents the first quartile (Q1)
It is the 25th percentile, meaning 25% of the data points are below this value.
Q1 is the median of the lower half of the dataset (excluding the overall median if the dataset size is odd).

c. Third Quartile (75%):
The upper edge of the box represents the third quartile (Q3)
It is the 75th percentile, meaning 75% of the data points are below this value.

* Interquartile Range (IQR)

The IQR measures the spread of the middle 50% of the data:

IQR = Q3-Q1

It shows the range between the 25th percentile (Q1) and the 75th percentile (Q2)

```python
# Boxplot :Show how the feature varies across classes, highlighting differences in central tendency (median), spread (IQR), and outliers.

sns.boxplot(data=iris_df, x='target', y='petal length (cm)')
plt.show()
```

# Method 4 : Violin Plot 

A violin plot is a data visualization technique that combines aspects of a box plot and a kernel density plot. 
It provides a comprehensive view of the distribution of a dataset, showing not only the summary statistics (like median and quartiles) but also the density of the data at different values. 
Violin plots are particularly useful for comparing distributions across different categories.



Key Features of Violin Plots

1. "Density Estimation":
   - The main feature of a violin plot is the density estimation, which is typically represented as a smoothed curve on either side of the central axis. This curve shows the distribution of the data points, indicating where data points are concentrated.
   - The width of the violin at any given y-value represents the density of the data points at that value. Wider sections indicate a higher density of data points.

2. "Summary Statistics":
   - Violin plots often include summary statistics such as:
     - "Median": A line in the center of the violin indicates the median of the data.
     - "Interquartile Range (IQR)": A box may be drawn around the median to represent the IQR, similar to a box plot.
     - "Whiskers": Some implementations may include whiskers that extend to show the range of the data.
     - This makes violin plots a superset of boxplots.

3. "Comparison Across Categories":
   - Violin plots are particularly effective for comparing distributions across multiple categories. Each category can be represented by a separate violin, allowing for easy visual comparison of their distributions.

4. "Orientation":
   - Violin plots can be oriented either vertically or horizontally, depending on the data being visualized and the preferences of the analyst.

```python
sns.violinplot(data=iris_df, x='target', y='petal length (cm)', palette='Set2', inner='box', scale='width')
plt.show()
```

Conclusions from the Violin Plot for `Petal Length` by Target

 "1. Clear Separation for Target `0` (Setosa):"
- The "density distribution" for target `0` is completely non-overlapping with the distributions for targets `1` (Versicolor) and `2` (Virginica).

- "Implication:"
  - Petal length is an excellent feature for separating target `0` from the other two classes.



 "2. Overlap Between Target `1` (Versicolor) and Target `2` (Virginica):"

- The density of target `1` is concentrated in the mid-range, while target `2` extends toward higher values.
- "Implication:"
  - Petal length alone cannot fully separate targets `1` and `2`.
  - Additional features (e.g., petal width or sepal dimensions) may help distinguish between these classes.



 "3. Increasing Median and Variability:"
- The "median petal length" increases significantly across the three targets:
  - Target `0` has the smallest median.
  - Target `1` has a moderate median.
  - Target `2` has the largest median.
- The "spread (variability)" also increases from target `0` to target `2`:
  - Target `0` has a very narrow and dense distribution.
  - Target `1` has a slightly wider spread.
  - Target `2` has the widest spread, indicating greater diversity in petal length within this class.
- "Implication:"
  - This systematic increase in both central tendency and spread aligns with class hierarchy in the Iris dataset.



 "4. Shape of Distributions (Density):"
- "Target `0` (Setosa):"
  - The density curve is compact and unimodal, reflecting uniform and small petal lengths.
- "Target `1` (Versicolor):"
  - The density curve is wider and slightly skewed, with a peak near the middle range of petal lengths.
- "Target `2` (Virginica):"

- "Implication:"

 "5. Class Hierarchy and Discriminative Power:"
- "Petal Length as a Feature:"

  - However, it is less effective in separating targets `1` (`Versicolor`) and `2` (`Virginica`) due to the overlapping regions in their distributions.
