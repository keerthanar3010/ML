# Features selection using Scatter and Pairplot for numerical features

(matplotlib.pyplot.scatter) A scatter plot is a graph that shows the relationship between two variables. It can be used to identify trends, patterns, and clusters in the data. Scatter plots can also help determine if one variable can predict another.

```python
!pip install matplotlib pandas numpy seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
import pandas as pd

#load the dataset
iris = load_iris()
```

```python
# Convert to pandas DataFrame
iris_df = pd.DataFrame(data=iris.data,columns=iris.feature_names)

# Add the target column
iris_df['target'] = iris.target

# Display the few rows
print(iris_df.tail())
```

```python
iris_df["sepal length (cm)"]
```

```python
# How many data points are present for each class/target
iris_df["target"].value_counts()

# iris is a balanced dataset as the number of the data points for each class/target is same
```

```python
#scatter plot of sepal length (cm) vs target

plt.scatter(iris_df["sepal length (cm)"],iris_df["target"], c='red',cmap='plasma')
plt.show()
```

- If c is a list of actual color names or hex codes, cmap is ignored.
- cmap='plasma': Maps colors values to the "plasma" colormap.

Common Colormaps :

'viridis' (default, perceptually uniform)

'plasma' (bright and high contrast)

'inferno' (dark background, good contrast)

'magma' (dark, warm colors)

'coolwarm' (good for diverging data)

```python
#scatter plot of sepal width (cm) vs target

plt.scatter(iris_df["sepal width (cm)"],iris_df["target"], c='yellow',cmap='plasma')
plt.show()
```

```python
#scatter plot of petal length (cm) vs target

plt.scatter(iris_df["petal length (cm)"],iris_df["target"], c='black',cmap='plasma')
plt.show()
```

```python
#scatter plot of petal width (cm) vs target

plt.scatter(iris_df["petal width (cm)"],iris_df["target"], c='green',cmap='plasma')
plt.show()
```

# PairPlot

```python
# Use pair plots to analyze relationships between all features and the target

# Pairwise scatter plots with hue as target
sns.pairplot(iris_df, hue='target', diag_kind='kde', palette='Set2')
plt.show()
```

# Compute correlations to quantify relationships

```python
correlations = iris_df.corr()
print(correlations['target'])
```

```python
# Select features with strong correlation

selected_features = ["petal length (cm)",'petal width (cm)']
new_df = iris_df[selected_features + ['target']]
new_df
```
