```python
from xgboost import XGBRegressor
# reference : https://xgboost.readthedocs.io/en/stable/python/python_api.html#module-xgboost.sklearn
```

```python
model = XGBRegressor(n_estimators=1500)
```

#### import dumy data from sklearn

```python
from sklearn.datasets import make_regression
```

```python
X,Y = make_regression(n_samples=2000,n_features=10,n_informative=7,noise=2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

#### train and test the model

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```
