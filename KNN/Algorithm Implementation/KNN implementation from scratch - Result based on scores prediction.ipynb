```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
score_df = pd.read_excel("./../Dataset/result_mathscience.xlsx")
score_df.describe()
```

```python
plt.scatter(score_df["Math"],score_df["Result"],c='blue',cmap='plasma')
```

```python
plt.scatter(score_df["Math"],score_df["Science"],c=score_df["Result"],cmap='plasma')
```

```python
X = score_df
X.shape[0]
```

```python
# Target value
target=(29,50)
```

```python
for i in range (0,X.shape[0]):
    point1 = (X["Math"][i],X["Science"][i])
    X["distance"][i] = math.dist(point1, target)
X = X.sort_values(by='distance')
print("X",X)
```
