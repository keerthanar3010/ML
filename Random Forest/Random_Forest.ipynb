# Random Forest

A Random Forest is called a meta-estimator because it combines the predictions of multiple individual decision tree models (estimators) to produce a final prediction. This process of combining multiple estimators to improve accuracy and reduce overfitting is characteristic of meta-estimators. 

##### Definition of parameters
n_estimators : The number of trees in the forest. (No. of base models/Decision tree)

#### Random Forest algorithm using sklearn library

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.33,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

##### Random forest with 10 base models

```python
model = RandomForestClassifier(n_estimators=10)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

##### Random forest with 50 base models

```python
model = RandomForestClassifier(n_estimators=50)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

When there are 50 base models, we have achieved 95.7% with test data. Now will increase the base models count to 100

##### Random forest with 100 base models

```python
model = RandomForestClassifier(n_estimators=100)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

Even after increasing the base models count from 50 to 100, score reamins same i.e., 95.7%

##### Random forest with 150 base models

```python
model = RandomForestClassifier(n_estimators=150)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

After increasing the base models count from 100 to 150, score reached to 96.8%

##### Random forest with 200 base models

```python
model = RandomForestClassifier(n_estimators=200)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
